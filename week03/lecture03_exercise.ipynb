```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Dataset: CO2 Emissions by Country 2000-2022
# Source: Our World in Data (https://ourworldindata.org/co2-emissions)
df = pd.read_csv('../data/co2_emissions.csv')
print(f"Loaded: {len(df)} rows | Countries: {df['Country'].nunique()} | Years: {df['Year'].min()}-{df['Year'].max()}")
print(df.head())
```

```python
# Explore before building

print("Countries:", df['Country'].unique())
print("\nCO2 range:", df['CO2_Mt'].min(), "to", df['CO2_Mt'].max(), "Mt")
print("\nRegional averages (2022):")
print(df[df['Year']==2022].groupby('Region')['CO2_Mt'].mean().sort_values(ascending=False).round(1))
```

## Task 1 — Multi-Series Line Chart with Highlight

**What to build:** A line chart showing CO2 emissions over time for **all Asian countries** in the dataset, with one country highlighted.

**Requirements:**
- All countries shown (for context), but only **one highlighted in colour** — your choice which
- All other lines in grey (#DDDDDD), thinner
- Highlighted country **labelled directly** at the end of its line (not in a legend)
- Insight title that names the highlighted country and its story

> 💡 `df[df['Region'] == 'Asia']` to filter; use `go.Figure()` with a loop for per-country control


```python
# Task 1 — Multi-series line with highlight
# YOUR CODE HERE
df_asia = df[df['Region'] == 'Asia']
countries = df_asia['Country'].unique()
highlight_country = 'China'

fig = go.Figure()

for country in countries:
    country_data = df_asia[df_asia['Country'] == country]

    if country == highlight_country:
        fig.add_trace(go.Scatter(
            x=country_data['Year'],
            y=country_data['CO2_Mt'],
            mode='lines',
            name=country,
            line=dict(color='#E74C3C', width=3)
        ))
    else:
        fig.add_trace(go.Scatter(
            x=country_data['Year'],
            y=country_data['CO2_Mt'],
            mode='lines',
            name=country,
            line=dict(color='#DDDDDD', width=1),
            showlegend=False
        ))

highlight_data = df_asia[df_asia['Country'] == highlight_country]

fig.add_annotation(
    x=highlight_data['Year'].max(),
    y=highlight_data[highlight_data['Year'] == highlight_data['Year'].max()]['CO2_Mt'].values[0],
    text=highlight_country,
    showarrow=False,
    font=dict(size=12, color='#E74C3C'),
    xanchor='left'
)

fig.update_layout(
    title={'text': "<b>China's Dominance:</b> CO2 emissions have tripled since 2000, far outpacing regional peers",
        'x': 0.05,
        'xanchor': 'left'},
    xaxis_title="Year",
    yaxis_title="CO₂ Emissions (Mt)",
    showlegend=False
)

fig.show()
```

## Task 2 — Slopegraph: Regional Change 2000 vs 2022

**What to build:** A slopegraph comparing **average regional CO2 emissions** between 2000 and 2022.

**Requirements:**
- One line per region (not per country — aggregate first)
- Colour: regions that increased = one colour; decreased = another
- Values labelled at both ends of each line
- No y-axis tick labels (the endpoint labels make them redundant)
- Insight title stating which regions moved most

> 💡 `df.groupby(['Region','Year'])['CO2_Mt'].mean().reset_index()` then filter to 2000 and 2022


```python
# Task 2 — Slopegraph: regional averages
# YOUR CODE HERE
df_slope = df.groupby(['Region', 'Year'])['CO2_Mt'].mean().reset_index()
df_slope = df_slope[df_slope['Year'].isin([2000, 2022])]
regions = df_slope['Region'].unique()

fig = go.Figure()

for region in regions:
    reg_data = df_slope[df_slope['Region'] == region].sort_values('Year')

    val_2000 = reg_data[reg_data['Year'] == 2000]['CO2_Mt'].values[0]
    val_2022 = reg_data[reg_data['Year'] == 2022]['CO2_Mt'].values[0]

    line_color = "seagreen" if val_2022 < val_2000 else "crimson"

    fig.add_trace(go.Scatter(
        x = reg_data['Year'], y = reg_data['CO2_Mt'],
        mode = 'lines+markers+text',
        name = region,
        line = dict(color = line_color, width = 3),
        marker = dict(size = 10),
        text=[f"{region} {val_2000:.0f}", f"{val_2022:.0f}"],
        textposition=["middle left", "middle right"],
        hoverinfo='none'
    ))

fig.update_layout(
    title="<b>Regional Shifts:</b> Asia and Africa see sharp increases while North America declines",
    xaxis=dict(
        type='category', 
        range=[-0.8, 1.8], # Give space for text labels
        showgrid=False
    ),
    yaxis=dict(
        showticklabels=False, # Requirements: "No y-axis tick labels"
        showgrid=False,
        zeroline=False,
        range=[df_slope['CO2_Mt'].min() - 500, df_slope['CO2_Mt'].max() + 500]
    ),
    plot_bgcolor='white',
    showlegend=False,
    height=900,
    margin=dict(l=200, r=200, t=100, b=100) # Extra margin for the long labels
)

fig.show()
```
